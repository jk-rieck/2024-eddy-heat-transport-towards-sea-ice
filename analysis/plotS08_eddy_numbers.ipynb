```python
import numpy as np
import xarray as xr
from matplotlib import pyplot as plt
plt.rcParams['figure.facecolor'] = 'white'
%config InlineBackend.print_figure_kwargs = {'bbox_inches': None}
from matplotlib.patches import Rectangle
```

```python
path = "../data_for_plots/" 
```

```python
eddy_numbers = xr.open_mfdataset(path + "eddy_lat_distribution.nc")
```

```python
fig, ax = plt.subplots(1)

eddy_numbers.number_cyc.plot(label="cyclones", color="limegreen", ls="-.")
eddy_numbers.number_anti.plot(label="anticyclones", color="firebrick", ls="-.")

plt.legend(loc=1, ncol=2)
plt.grid()
plt.title("Number of detected eddies")
plt.ylabel("number")
plt.xticks(np.arange(0, 3.2e6, 0.5e6), np.arange(0, 3200, 500))
plt.xlabel("y coordinate (km)")
plt.xlim(0, 3.2e6)
plt.ylim(-10, 6400)

ax.vlines(1.57e6, -10, 6400,  colors='gray', linestyle='--', alpha=0.2)
ax.add_patch(Rectangle((0, -10), 1.57e6, 6410, alpha=0.2, facecolor="gray", edgecolor=None))
ax.text(1.57e6, 50, r"$SI_{15} $", ha="right", fontsize=12, color="gray")
ax.vlines(1.20e6, -10, 6400,  colors='dimgray', linestyle='--', alpha=0.2)
ax.add_patch(Rectangle((0, -10), 1.20e6, 6410, alpha=0.2, facecolor="dimgray", edgecolor=None))
ax.text(1.20e6, 50, r"$SI_{80} $", ha="right", fontsize=12, color="dimgray")

ax.add_patch(Rectangle((3.07e6, -10), 0.13e6, 6410, alpha=0.4, facecolor="skyblue", hatch="OO", edgecolor="k"));

plt.subplots_adjust(bottom=0.25)

plt.savefig("figS08-number_eddies.png", dpi=300)
```

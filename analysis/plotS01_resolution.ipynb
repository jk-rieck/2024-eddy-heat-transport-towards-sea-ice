```python
import numpy as np
import xarray as xr
from matplotlib import pyplot as plt
import cmocean.cm as cmo
plt.rcParams['figure.facecolor'] = 'white'
%config InlineBackend.print_figure_kwargs = {'bbox_inches': None}
from mpl_toolkits.axes_grid1 import make_axes_locatable
```

```python
path = "../data_for_plots/" 
```

```python
Ro = xr.open_dataset(path + "meanRossbyRadius.nc").Ro
```

```python
fig, ax = plt.subplots(1, 1, figsize=(24/5, 32/5))
divider = make_axes_locatable(ax)
cax = divider.append_axes("right", size="5%", pad=0.2)
   
a1 = ax.contourf(np.linspace(0, 2400, 240), np.linspace(0, 3200, 320), Ro/10000., 
                   levels=np.arange(0, 7), cmap=cmo.solar)
cbar = plt.colorbar(a1, cax=cax, label=r"$L_{d}/\Delta x$")
ax.set_title(r"$L_{d}/\Delta x$", fontsize=14)
ax.set_xticks(np.arange(0, 2400, 500))
ax.set_xlabel("x coordinate (km)")
ax.set_yticks(np.arange(500, 3100, 500))
ax.set_ylabel("y coordinate (km)")

c0 = ax.contour(np.linspace(0, 2400, 240), np.linspace(0, 3200, 320), Ro/10000., 
                levels=[2], colors="snow", linestyles="-", linewidths=3)
c1 = ax.contour(np.linspace(0, 2400, 240), np.linspace(0, 3200, 320), Ro/10000., 
                levels=[2], colors="dimgray", linestyles="--")

plt.subplots_adjust(left=0.2, right=0.8, top=0.8, bottom=0.2)

plt.savefig('figS01-Rossby_over_dx.png', dpi=300)
```

```python
import numpy as np
import xarray as xr
from matplotlib import pyplot as plt
import cmocean.cm as cmo
from matplotlib.patches import Rectangle
```

```python
path = "../data_for_plots/" 
```

```python
OIQ = xr.open_mfdataset(path + "ocean_ice_heat_transports_0201-01-01_0300-12-30_JJA_zonal_integral.nc")
VHT = xr.open_mfdataset(path + "vertical_heat_transports_0201-01-01_0300-12-30_JJA_zonal_integral.nc")
```

```python
fig, ax = plt.subplots(2, 1, figsize=(10, 5))
OIQ.total.plot(ax=ax[0], color="yellowgreen", label="total OIQ")
OIQ.prime.plot(ax=ax[0], color="steelblue", label="transient OIQ")
OIQ.prime_CME.plot(ax=ax[0], color="steelblue", label="CME OIQ", ls="--")
OIQ.prime_CME_anti.plot(ax=ax[0], color="firebrick", label="anticycl. OIQ", ls="--")
OIQ.prime_CME_cycl.plot(ax=ax[0], color="limegreen", label="cycl. OIQ", ls="--")
ax[0].legend(loc=5, bbox_to_anchor=(1.01, 0.7))
ax[0].grid()
ax[0].set_xlim(0, 3.2e6)
ax[0].set_ylim(-0.01, 0.17)
ax[0].set_title("Vertical HT")
ax[0].set_ylabel("HT (TW)")
ax[0].set_xticklabels([])
ax[0].vlines(1.57e6, -0.01, 0.17,  colors='gray', linestyle='--', alpha=0.2)
ax[0].add_patch(Rectangle((0, -0.01), 1.57e6, 0.18, alpha=0.1, facecolor="gray", edgecolor=None))
ax[0].vlines(1.20e6, -0.01, 0.17,  colors='dimgray', linestyle='--', alpha=0.2)
ax[0].add_patch(Rectangle((0, -0.01), 1.20e6, 0.18, alpha=0.1, facecolor="dimgray", edgecolor=None))
ax[0].add_patch(Rectangle((3.07e6, -0.01), 0.13e6, 0.18, alpha=0.4, facecolor="skyblue", hatch="OO", edgecolor="k"))


VHT.total.plot(ax=ax[1], color="yellowgreen", label="total VHT")
VHT.prime.plot(ax=ax[1], color="steelblue", label="transient VHT")
VHT.CME.plot(ax=ax[1], color="steelblue", label="CME VHT", ls="--")
VHT.CME_anti.plot(ax=ax[1], color="firebrick", label="anticycl. VHT", ls="--")
VHT.CME_cycl.plot(ax=ax[1], color="limegreen", label="cycl. VHT", ls="--")
VHT.DIFF.plot(ax=ax[1], color="tab:orange", label="convection/\ndiffusion", ls=":")
ax[1].legend(loc=5, bbox_to_anchor=(1.01, 0.6))
ax[1].grid()
ax[1].set_xlim(0, 3.2e6)
ax[1].set_ylim(-0.1, 1.1)
ax[1].set_title("")
ax[1].set_ylabel("HT (TW)")
ax[1].set_xticklabels([0, 500, 1000, 1500, 2000, 2500, 3000])
ax[1].set_xlabel("y-coordinate km")
ax[1].vlines(1.57e6, -0.1, 0.17,  colors='gray', linestyle='--', alpha=0.2)
ax[1].add_patch(Rectangle((0, -0.1), 1.57e6, 1.2, alpha=0.1, facecolor="gray", edgecolor=None))
ax[1].text(1.57e6, 0.95, r"$SI_{15} $", ha="right", fontsize=12, color="gray")
ax[1].vlines(1.20e6, -0.1, 0.17,  colors='dimgray', linestyle='--', alpha=0.2)
ax[1].add_patch(Rectangle((0, -0.1), 1.20e6, 1.2, alpha=0.1, facecolor="dimgray", edgecolor=None))
ax[1].text(1.20e6, 0.95, r"$SI_{80} $", ha="right", fontsize=12, color="dimgray")
ax[1].add_patch(Rectangle((3.07e6, -0.1), 0.13e6, 1.2, alpha=0.4, facecolor="skyblue", hatch="OO", edgecolor="k"))

plt.subplots_adjust(hspace=0.05)
fig.text(0.05, 0.83, "a)", fontsize=18)
fig.text(0.05, 0.45, "b)", fontsize=18)

plt.savefig("fig08-vertical_HT.png", dpi=300)
```
